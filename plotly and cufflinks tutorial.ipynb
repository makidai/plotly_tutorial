 - box
 - spread
 - ratio
 - heatmap

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100, 3), columns = ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 25
df['B'] = df['B'].cumsum() + 25
df['C'] = df['C'].cumsum() + 25
```

```python
df.shape
```

```python
df.iplot()
```

```python
plt.plot(df)
```

```python
df.plot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 15)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle = 'Sex', yTitle = 'Survived')
```

```python
titanic['sex'].value_counts()
```

```python
df.iplot(kind = 'bar', bargap=0.5, barmode='stack')
```

```python
df.iplot(kind = 'barh', bargap=0.5, barmode='stack')
```

```python
cf.getThemes()
```

```python
cf.set_config_file(theme='polar')
df.iplot(kind = 'bar', bargap=0.5, barmode='stack')
```

```python
df.iplot(kind = 'box')
```

```python
df.iplot(kind = 'area', fill=True)
```

```python
df = pd.DataFrame({'X': [10, 20, 30, 20 ,10], 'Y': [10, 20, 30, 20 ,10], 'Z': [10, 20, 30 ,20, 10]})
```

```python
df.iplot(kind='surface', colorscale='rdylbu')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind='surface')
```

```python
cf.datagen.scatter3d(2, 150, mode='stocks').iplot(kind='scatter3d', x='x', y='y',z='z', size=5)
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind='hist', bins=25, barmode='overlay', bargap=0.5)
```

```python
cf.datagen.bubble3d(5,4,mode='stocks').iplot(kind='bubble3d', x='x',y='y',z='z', size='size')
```

```python
cf.datagen.heatmap(20,20).iplot(kind='heatmap', colorscale='spectral', title='Cufflinks - Heatmap')
```
